```python
#suma de matrices
import numpy as np

n=int(input("Ingrese el numero de filas: "))
m=int(input("Ingrese el numero de columnas: "))

A=[]
B=[]

print("Ingrese los elementos de la matriz A: ")
for i in range(n):
    fila= list(map(int, input(f"fila{i+1}:").split(","))) #.split para detectar los espacios oneste caso "," entre los numeros
    A.append(fila)

print("Ingrese los elementos de la matriz B: ")
for i in range(n):
    fila= list(map(int, input(f"fila{i+1}:").split(",")))
    B.append(fila)

A= np.array(A)
B= np.array(B)

suma= A+B
print("matriz A es:", A)

print("la matriz b es:",B)
print("La suma de las matrices es: ")
print(A+B)
```
