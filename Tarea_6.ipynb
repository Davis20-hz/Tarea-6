```python
import numpy as np

A = np.array([
    [11, 7, -1],
    [-3, 5, 2],
    [4, 2, -2]
])

B = np.array([
    [1, 1, 1],
    [10, 3, -5],
    [-6, -10, -3]
])
print("Matriz A:", A)
print("Matriz B:", B)

# Calcular 2B + 3A
dosB = 2 * B
tresA = 3 * A

resultado1 = dosB + tresA
print("Resultado de 2B + 3A:\n", resultado1)

# (B + 2A)^2
dosA = 2 * A          # Multiplicación escalar
C = B + dosA          # Suma de matrices: C = B + 2A

resultado2 = C @ C    # @ Producto matricial (C^2) o np.linalg.matrix_power(A, 2)
print("Resultado de (B + 2A)^2:\n", resultado2)

# A^2 + 3B
A2 = np.linalg.matrix_power(A, 2)            # Producto matricial A^2
tresB = 3 * B         # Multiplicación escalar 3B
resultado3 = A2 + tresB
print("Resultado de A^2 + 3B:\n", resultado3) #\n para dejar espacio
```

```python
import numpy as np

def crear_matriz_cuadrada():
    while True:
        n = int(input("Ingrese el número de filas (3 a 5): "))
        m = int(input("Ingrese el número de columnas (3 a 5): "))

        if n == m and 3 <= n <= 5:
            break
        else:
            print("La matriz no es cuadrada. Intente nuevamente.\n")

    #   Matriz con números aleatorios del 1 al 20
    T = np.random.randint(1, 21, (n, m))
    print("\nMatriz generada T:\n", T, "\n")

    #Primera fila al revés
    print("Primera fila en orden inverso:", T[0, ::-1])

    #Primera columna en orden inverso
    print("Primera columna en orden inverso:", T[::-1, 0])

    #Primer y último elemento
    print("Primer elemento:", T[0, 0])
    print("Último elemento:", T[-1, -1])

    #Máximo, mínimo, promedio de la diagonal, traza
    diagonal = np.diag(T)
    print("\nDiagonal:", diagonal)
    print("Máximo de la diagonal:", diagonal.max())
    print("Mínimo de la diagonal:", diagonal.min())
    print("Promedio de la diagonal:", diagonal.mean())
    print("Traza (suma de la diagonal):", diagonal.sum())

    #Suma de valores de la primera fila + última columna
    suma_primera_fila = T[0, :].sum()
    suma_ultima_columna = T[:, -1].sum()
    print("\nSuma de la primera fila:", suma_primera_fila)
    print("Suma de la última columna:", suma_ultima_columna)
    print("Suma total (primera fila + última columna):",
          suma_primera_fila + suma_ultima_columna)

    # Suma de elementos pares recorriendo las filas

    pares = T[T % 2 == 0]
    print("\nSuma de elementos pares:", pares.sum())

    # Suma de elementos impares recorriendo las columnas

    impares = T[T % 2 == 1]
    print("Suma de elementos impares:", impares.sum())

    # Promedios de filas y columnas

    print("\nPromedio de cada fila:", T.mean(axis=1))
    print("Promedio de cada columna:", T.mean(axis=0))

    #Antidiagonal y su promedio

    antidiagonal = np.fliplr(T).diagonal()
    print("\nAntidiagonal:", antidiagonal)
    print("Promedio de la antidiagonal:", antidiagonal.mean())

crear_matriz_cuadrada()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear el rango de valores de x

x = np.linspace(-5, 5, 500) #500 pareciar mejor los valores

def solucion(x, C):
    return (x - 1) + C * np.exp(-x)

# Valores para C:

Cs = [2, 0, -2]
colores = ["blue", "green", "red"]
etiquetas = ["C > 0", "C = 0", "C < 0"]

# Crear el gráfico

plt.figure(figsize=(8, 5))

# Graficar cada curva según su valor de C
for C, color, label in zip(Cs, colores, etiquetas):
    y = solucion(x, C)
    plt.plot(x, y, color=color, label=label)


plt.title("Familia de soluciones de y = (x-1) + Ce^{-x}")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (3*x + 7) / (x**2 - 4)

# Rango de x, evitando valores donde se divide por 0
x = np.linspace(-10, 10, 1500)

# Eliminar los puntos donde x se acerca demasiado a ±2
x_filtered = x[(np.abs(x - 2) > 0.02) & (np.abs(x + 2) > 0.02)]

# Calcular f(x) usando el vector filtrado
y = f(x_filtered)

plt.figure(figsize=(9, 6))

plt.plot(x_filtered, y, label="f(x) = (3x + 7)/(x^2 - 4)", color="blue")

#Asíntotas verticales x = 2 y x = -2
plt.axvline(2, color="red", linestyle="--", label="Asíntotas verticales")
plt.axvline(-2, color="red", linestyle="--")

#Asíntota horizontal y = 0
plt.axhline(0, color="green", linestyle="--", label="Asíntota horizontal")

# imitar el eje
plt.ylim(-10, 10)

plt.title("Gráfica de f(x) = (3x + 7)/(x^2 - 4)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
